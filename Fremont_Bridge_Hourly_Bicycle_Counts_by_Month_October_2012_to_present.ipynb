```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Fremont_Bridge_Hourly_Bicycle_Counts_by_Month_October_2012_to_present.csv", index_col = 'Date', parse_dates = True)
df.head()
```

```python
plt.plot(df)
plt.title("Bicyle Trips")
plt.ylabel("Number of Trips")
plt.show() #plot below shows the number of bike trips over the bridge, and its clearly visible that during winter 
           #there were less bicycle trips compare to summers 
```

```python
df.plot(kind = 'scatter', x = 'Fremont Bridge West Sidewalk', y = 'Fremont Bridge East Sidewalk')
df.plot(kind = 'line')
plt.show()
```

```python
df.resample('W').sum().plot() #Plot showed by weekly
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Fremont Bridge West Sidewalk'], color='purple')
plt.ylabel('Number of Trips')
plt.title('Bicyle trips over the West sidewalk bridge')

plt.subplot(1, 2, 2)
plt.plot(df['Fremont Bridge East Sidewalk'], color='green')
plt.ylabel('Number of Trips')
plt.title('Bicyle Trips over the East sidewalk Bridge')
plt.show()
```

```python
#Plot above shows  East Sidewalk bridge has slightly more Bicycle traffic then west Sidewalk Bridge
```
